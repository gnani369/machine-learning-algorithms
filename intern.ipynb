```python
# Imputations

other_skills = data['Other_skills'].mode()[0]
data['Other_skills'] = data['Other_skills'].replace(np.NaN,other_skills) 

degree = data['Degree'].mode()[0]
data['Degree'] = data['Degree'].replace(np.NaN,degree)

stream = data['Stream'].mode()[0]
data['Stream'] = data['Stream'].replace(np.NaN,stream)

data['Performance_PG'] = data['Performance_PG'].replace(np.NaN,0)

ug = data['Performance_UG'].mode()[0]
data['Performance_UG'] = data['Performance_UG'].replace(np.NaN,ug)

data['Performance_12'] = data['Performance_12'].replace(np.NaN,0)

ssc = data['Performance_10'].mode()[0]
data['Performance_10'] = data['Performance_10'].replace(np.NaN,ssc)
```

```python
#optimum number of clusters using elbow curve
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(data) 
    
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 8))
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
